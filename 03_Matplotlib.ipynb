# MathPlotlib

## Part One

```python
import matplotlib.pyplot as plt
import numpy as np

# or show()
%matplotlib inline 
```

```python
x =np.linspace(0,5,11)
y=x**2
```

### Functional method

```python
plt.plot(x,y, 'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y=x**2')
plt.legend('y')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'g')
```

### Object oriented

```python
fig=plt.figure()

axes=fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, widthg, height 
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('y = x**2')

axes2=fig.add_axes([0,0.5,0.4,0.4])
axes2.plot(y,x, 'r')
```

## Part Two

```python
fig, axes=plt.subplots(nrows=1, ncols=2)

# axes are iterable, for a in axes: ...

axes[0].plot(x,x,'r', label='y=x')


axes[1].plot(x,y,'b', label='y=x**2')
axes[1].plot(x,y*2,'g', label='y=2*(x**2)')

axes[0].legend(loc=(0.2,0.8))
axes[1].legend(loc='center') # location codes in documentation


plt.tight_layout()
```

### Figure Size and DPI

```python
fig=plt.figure(figsize=(3,2), dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

### Save figure to the file

```python
fig.savefig('my_pict.jpg') # .jpg, .png, .pdf ...
```

## Part Three

```python
fig=plt.figure(figsize=(10,6))

ax=fig.add_axes([0,0,1,1])


ax.plot(x,x+1,color='#cc1b97', lw=5)
ax.plot(x,x+2,color='orange', marker='o', markersize=30, markerfacecolor='#cc1b97', markeredgewidth=5, markeredgecolor='#461282')
ax.plot(x,x+3, color='purple', linestyle='dotted')
ax.plot(x,x+4, color='blue', ls='--', lw=10)

ax.plot(x,-x+10,color='#32a893', linewidth=60, alpha=0.7)

ax.set_xlim([1,5])
ax.set_ylim([1,10])
```
